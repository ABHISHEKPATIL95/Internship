```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv('https://github.com/dsrscientist/IBM_HR_Attrition_Rate_Analytics/raw/master/ibm-hr-analytics-employee-attrition-performance.zip', compression='zip')
```

```python
# Explore the data
print(data.head())
print(data.info())
```

```python
# Calculate the attrition rate
attrition_rate = data['Attrition'].value_counts(normalize=True) * 100
print(f"Attrition Rate:\n{attrition_rate}")
```

```python
sns.countplot(x=data['Attrition'])
plt.title('Attrition Distribution')
plt.show()
```

```python
# Explore attrition by different factors
sns.countplot(data=data, x='Attrition', hue='Department')
plt.title('Attrition by Department')
plt.show()
```

```python
sns.countplot(data=data, x='Attrition', hue='JobRole')
plt.title('Attrition by Job Role')
plt.show()
```
